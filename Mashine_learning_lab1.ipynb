###3. Построить pairplot для набора данных.

```python
sns.pairplot(data, hue='alcohol')
plt.show()
```

###4. Выбрать не менее 2-х признаков с неким распределением значений, и отобразить это распределение с помощью гистограмм hist и kdeplot, jointplot.

```python
plt.title('Распределение количества качества вина по количеству остаточного сахара')
sns.histplot(data = data, y = 'quality', x = 'residual sugar')
plt.ylabel('Качество вина')
plt.xlabel('Количество остаточного сахара')
```

```python
plt.title('Распределение качества вина по проценту алкоголя')
sns.kdeplot(data = data, x = 'quality', y = 'alcohol', fill = True)
plt.ylabel('Процент алкоголя')
plt.xlabel('Качество вина')
```

```python
arrData = data[['citric acid', 'free sulfur dioxide', 'quality']]
sns.jointplot(arrData, x = 'quality', y = 'free sulfur dioxide', hue = 'citric acid')
plt.title('Распределение качества вина и содержания свободного диоксида серы по количеству лимонной кислоты', pad = 90)
```

###5. Выбрать часть признаков датасета и отобразить корреляцию между ними в виде тепловой карты (heatmap).

```python
heat = data[['fixed acidity',	'volatile acidity',	'citric acid',	'quality']]
heat_cor = heat.corr()
plt.figure()
plt.title('Корреляция между качеством вина, фиксированной кислотностью, летучей кислотностью и содержанием лимонной кислоты', pad = 20)
sns.heatmap(heat_cor, annot = True, linewidths = 0.3)
```

###6. Выбрать 3 признака (имеющие некое распределение значений), целевую переменную (категориальную) и отобразить для них графики размаха (ящик с усами, box plot).

```python
sns.boxplot(x='quality', y='citric acid', data=data)
plt.title('Box Plot for Citric Acid')

plt.show()
```

```python
sns.boxplot(x='quality', y='residual sugar', data=data)
plt.title('Box Plot for Residual Sugar')

plt.show()
```

```python
sns.boxplot(x='quality', y='alcohol', data=data)
plt.title('Box Plot for Alcohol')

plt.show()
```

###7. Визуализировать некоторые статистики, для разных атрибутов с использованием следующих инструментов: violinplot, stripplot, swarmplot, pie.

```python
sns.violinplot(x='quality', y='alcohol', data=data)
plt.title('Violin plot for Alcohol')

plt.show()
```

```python
sns.stripplot(x='quality', y='residual sugar', data=data)
plt.title('Stripplot for residual sugar')

plt.show()
```

```python
sns.swarmplot(x='quality', y='sulphates', data=data)
plt.title('Swarm Plot for Sulphates and Quality')

plt.show()
```

```python
quality_counts = data['quality'].value_counts().sort_index()
labels = [f'Quality {quality}: {count}' for quality, count in zip(quality_counts.index, quality_counts)]

plt.figure(figsize=(8, 8))
plt.pie(quality_counts, labels=labels, autopct='%1.2f%%')
plt.title('Pie Chart for Wine Quality')

plt.show()
```
